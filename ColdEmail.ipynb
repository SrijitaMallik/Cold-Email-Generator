```python
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_groq import ChatGroq

# JSON parser
json_parser = JsonOutputParser()

# Stronger prompt
prompt = PromptTemplate(
    template="""
    You are an expert HR assistant. 
    From the given job description, extract the following fields with maximum detail:

    - role (use the full official job title, not abbreviated)
    - experience (mention exact years if written, else summarize qualifications)
    - skills (return as a list, include ALL technologies, tools, practices mentioned)
    - description (summarize the job in 3–4 sentences, keeping important context)

    The output must be ONLY JSON in this format:
    {format_instructions}

    Job Description:
    {job_text}
    """,
    input_variables=["job_text"],
    partial_variables={"format_instructions": json_parser.get_format_instructions()},
)

# Groq model
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0,
    groq_api_key="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"
)

# Build pipeline
chain = prompt | llm | json_parser

# ✅ Full job description (not just 1 line!)
job_text = """
We are looking for a Software Engineer II to design and implement core components of our data science services and platform and scale it to serve Nike globally. 
The focus of this engineer will be at the intersection of computer systems and machine learning. 
You will also work collaboratively with the data science and analytics teams to ship scalable, performant and reliable solutions.

Requirements:
- 2+ years professional software development experience
- Strong Computer Science fundamentals
- Programming experience with at least one modern language such as Java, Python, Golang
- Experience with Infrastructure as Code / DevOps practices and technologies like Terraform
- Experience with Cloud services (e.g. AWS Cloud EC2, S3, DynamoDB, Azure Cloud, etc.)
- Experience developing and using microservices
"""

# Run
result = chain.invoke({"job_text": job_text})

print(result)
```

```python
import pandas as pd
import uuid
import chromadb

# Load portfolio CSV
df = pd.read_csv(r"C:\Users\KIIT\Downloads\my_portfolio.csv")

# Initialize ChromaDB
client = chromadb.PersistentClient(path="vectorstore")
collection = client.get_or_create_collection(name="portfolio")

# Insert if collection empty
if collection.count() == 0:
    for _, row in df.iterrows():
        collection.add(
            documents=[row["Techstack"]],
            metadatas=[{"links": row["Links"]}],
            ids=[str(uuid.uuid4())]
        )

print(f"✅ Inserted {collection.count()} documents into ChromaDB")

# Example job (replace with your parsed job data)
job = {
    "role": "Software Engineer II, AI/ML Platforms",
    "experience": "2+ years professional software development experience",
    "skills": ["Python", "AWS", "Terraform", "Microservices"],
    "description": "We are looking for a Software Engineer II..."
}

# Query portfolio against job skills
links = collection.query(
    query_texts=job['skills'],  # list of skills
    n_results=2
).get('metadatas', [])

print("🔗 Recommended portfolio links:", links)
```
